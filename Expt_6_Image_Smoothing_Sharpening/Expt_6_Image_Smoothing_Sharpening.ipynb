# Image Smoothing And Sharpening

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("kalin.jpg")
```

```python
img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(img, cmap='gray')
plt.imsave('imgGray.jpg', img)
plt.axis('off')
```

```python
import os
size = os.path.getsize("imgGray.jpg")
print("Size: ", size, "Bytes")
print("Rows:", img.shape[0], "Columns:", img.shape[1])
```

# Image Smoothing

```python
n = int(input("Enter Kernel Size: "))
```

```python
kernel1 = np.ones((n, n), np.float32) / (n*n)
```

```python
kernel1
```

```python
imgBlur = cv2.filter2D(img, ddepth = -1, kernel = kernel1)
plt.imshow(imgBlur, cmap='gray')
plt.axis('off')
```

## Averaging Filter 

```python
AvgBlur = cv2.boxFilter(img, -1, (n, n), normalize=True)  
```

```python
plt.imshow(AvgBlur, cmap='gray')
```

```python
hs = np.hstack((img, AvgBlur))
plt.figure(figsize = (20,8))
plt.imshow(hs, cmap='gray')
plt.axis('off')
```

## Gaussian Blur

```python
GuasBlur = cv2.GaussianBlur(img, (n,n), sigmaX= 0, sigmaY=0) 
```

```python
plt.imshow(GuasBlur, cmap='gray')
plt.axis('off')
```

```python
hs = np.hstack((img, GuasBlur))
plt.figure(figsize = (20,8))
plt.imshow(hs, cmap='gray')
plt.axis('off')
```

## Median Blur

```python
MedBlur = cv2.medianBlur(img, n)
```

```python
plt.imshow(MedBlur, cmap='gray')
```

```python
hs = np.hstack((img, MedBlur))
plt.figure(figsize = (20,8))
plt.imshow(hs, cmap='gray')
plt.axis('off')
```

## Bilateral Filtering

```python
BilBlur = cv2.bilateralFilter(img, d=9, sigmaColor=75, sigmaSpace=75)
```

```python
plt.imshow(BilBlur, cmap='gray')
```

```python
hs = np.hstack((img, BilBlur))
plt.figure(figsize = (20,8))
plt.imshow(hs, cmap='gray')
plt.axis('off')
```

# Image Sharpening

```python
# kernel for image sharpening
kernel2 = np.array([[0, -1,  0],
                    [-1,  5, -1],
                    [0, -1,  0]])
```

```python
imgBlur = cv2.imread("kalinBlur.jpg", 0)
```

```python
plt.imshow(imgBlur, cmap='gray')
```

```python

imgSharp = cv2.filter2D(imgBlur, ddepth=-1, kernel=kernel2)
```

```python
plt.imshow(imgSharp, cmap='gray')
```

```python
hs = np.hstack((imgBlur, imgSharp))
plt.figure(figsize = (20,8))
plt.imshow(hs, cmap='gray')
plt.axis('off')
```
