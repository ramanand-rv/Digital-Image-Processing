# Image Transformation

```python
# importing dependencies
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# reading image
img = cv2.imread("guddu.jpg")
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.axis('off')
plt.imshow(img, cmap='gray')
```

```python
print("Dimensions: ", img.shape)
```

# Image Scaling

```python
# Down scaling
def scaleDown(img):
  print("------------Scaling Down Image------------")
  x = int(input("Enter Scaling Percentage: "))
  print("Original Image Dimension: ", img.shape)

  row = int(img.shape[0] * (100 - x)/ 100)
  col = int(img.shape[1] * (100 - x)/ 100)

  dimension = (col, row)

  imgDown = cv2.resize(img, dimension, interpolation = cv2.INTER_AREA)
  print("Dimensions of Scaled Down Image: ", imgDown.shape)
  return imgDown
```

```python
imgDown = scaleDown(img)
plt.axis('off')
plt.imshow(imgDown, cmap = 'gray')
```

```python
# Up scaling
def scaleUp(img):
  print("------------Scaling Up Image------------")
  x = int(input("Enter Scaling Factor: "))
  print("Original Image Dimension: ", img.shape)

  row = int(img.shape[0] * (x))
  col = int(img.shape[1] * (x))

  dimension = (col, row)

  imgUp = cv2.resize(img, dimension, interpolation = cv2.INTER_AREA)
  print("Dimensions of Scaled Down Image: ", imgUp.shape)
  return imgUp
```

```python
imgUp = scaleUp(img)
plt.axis('off')
plt.imshow(imgUp, cmap = 'gray')
```

# Thresholding

```python
# Thresholding
def Threshold(img):
  print("-----------------Image Threshold-----------------")
  thresh = np.float32(input("Enter Threshold Value: "))
  
  thresh, imgThresh = cv2.threshold(img, thresh, 255, cv2.THRESH_BINARY)

  return imgThresh
```

```python
imgThresh = Threshold(img)
plt.axis('off')
plt.imshow(imgThresh, cmap='gray')
```

# Image Translation

```python
# image Translation
def Translation(img):
  print("-------------Image Translation-------------")
  tx = int(input("Enter value of Tx: "))
  ty = int(input("Enter value of Ty: "))

  row, col = img.shape

  # TransMatrix = np.float32([[1, 0, 0], [0, 1, 0], [tx, ty, 0]])
  TransMatrix = np.float32([[1, 0, tx], [0, 1, ty]])
  # print(TransMatrix)


  imgTrans = cv2.warpAffine(img, TransMatrix, dsize = (col, row))

  return imgTrans
```

```python
imgTrans = Translation(img)
plt.axis('off')
plt.imshow(imgTrans, cmap = 'gray')
```

# Image Rotation

```python
# Image Rotation
def Rotation(img):
  print("--------------Image Rotation--------------")
  theta = int(input("Enter Angle: "))
  row, col = img.shape

  # RotMatrix = np.float32([[np.cos(theta), - np.sin(theta)], [np.sin(theta), np.cos(theta)]])
  RotMatrix = cv2.getRotationMatrix2D((col/2, row/2), theta, 1)

  imgRot = cv2.warpAffine(img, RotMatrix, (col, row))

  return imgRot
```

```python
imgRot = Rotation(img)
plt.axis('off')
plt.imshow(imgRot, cmap = 'gray')
```

# Shear Vertical and Horizontal

```python
# shear
def verShear(img):
  print("-------------Vertical/Horizontal Shear-------------")
  sv = np.float32(input("Enter Shear Vertical Value: "))
  sh = np.float32(input("Enter Shear Horizontal Value: "))
  row, col = img.shape

  verMatrix = np.float32([[1, sh, 0], [sv, 1, 0], [0, 0, 1]])


  imgVer = cv2.warpPerspective(img, verMatrix, (int(col*1.2), int(row*1.2)))

  return imgVer
```

```python
imgVer = verShear(img)
plt.axis('off')
plt.imshow(imgVer, cmap = 'gray')
```
