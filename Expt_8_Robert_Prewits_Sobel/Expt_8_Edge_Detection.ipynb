## Edge Detection

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
```

```python
img = cv2.imread("dhindora.jpg", 0)
img = cv2.GaussianBlur(img, (7, 7), 0)
plt.imshow(img, cmap='gray')
plt.axis('off')
```

## Robert Filter

```python
robertx = np.array([[1, 0], [0, -1]])

roberty = np.array([[0, 1], [-1, 0]])
```

```python
x = ndimage.convolve( img, robertx )
y = ndimage.convolve( img, roberty )

imgR = np.sqrt( np.square(y) + np.square(x))
```

```python
from PIL import Image
```

```python
imgRobert = Image.fromarray( np.asarray( np.clip(imgR, 0, 255), dtype="uint8"), "L" )
```

```python
plt.imshow(imgRobert, cmap='gray')
plt.axis('off')
```

## Prewitts Filter

```python
prex = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])

prey = np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]])
```

```python
x = ndimage.convolve( img, prex )
y = ndimage.convolve( img, prey )

imgP = np.sqrt( np.square(y) + np.square(x))
```

```python
imgPre = Image.fromarray( np.asarray( np.clip(imgP, 0, 255), dtype="uint8"), "L" )
```

```python
plt.imshow(imgPre, cmap='gray')
plt.axis('off')
```

## Sobel Filter

```python
sobelx = cv2.Sobel(img, ddepth=cv2.CV_64F, dx=1, dy=0, ksize=7) 

sobely = cv2.Sobel(img, ddepth=cv2.CV_64F, dx=0, dy=1, ksize=7) 
```

```python
sobelx = Image.fromarray( np.asarray( np.clip(sobelx, 0, 255), dtype="uint8"), "L" )
plt.imshow(sobelx, cmap='gray')
plt.axis('off')
```

```python
sobely = Image.fromarray( np.asarray( np.clip(sobely, 0, 255), dtype="uint8"), "L" )
plt.imshow(sobely, cmap='gray')
plt.axis('off')
```

```python
sobelxy = cv2.Sobel(img, ddepth=cv2.CV_64F, dx=1, dy=1, ksize=7)
```

```python
imgSob = Image.fromarray( np.asarray( np.clip(sobelxy, 0, 255), dtype="uint8"), "L" )
plt.imshow(imgSob, cmap='gray')
plt.axis('off')
```
