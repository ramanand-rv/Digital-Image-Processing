## Aim: Study and plot histogram of different noise models

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("permanent.jpg", 0)
```

```python
plt.imshow(img, cmap='gray')
plt.axis('off')
```

```python
plt.hist(img.ravel(), 256, [0, 256])
plt.show()
```
