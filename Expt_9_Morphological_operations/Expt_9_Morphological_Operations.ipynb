## Aim: Perform Morphological operations.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("scared.jpg", 0)
plt.imshow(img, cmap='gray')
plt.axis('off')
```

## Erosion

```python
kernel = np.ones((5,5),np.uint8)

erosion = cv2.erode(img,kernel,iterations = 2)
```

```python
plt.imshow(erosion, cmap='gray')
plt.axis('off')
```

```python
imgero = np.hstack((img, erosion))
plt.imshow(imgero, cmap='gray')
plt.axis('off')
```

## Dilation

```python
dilation = cv2.dilate(img,kernel,iterations = 2)
```

```python
plt.imshow(dilation, cmap='gray')
plt.axis('off')
```

```python
imgero = np.hstack((img, dilation))
plt.imshow(imgero, cmap='gray')
plt.axis('off')
```

## Opening

```python
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
```

```python
plt.imshow(opening, cmap='gray')
plt.axis('off')
```

```python
imgero = np.hstack((img, opening))
plt.imshow(imgero, cmap='gray')
plt.axis('off')
```

## Closing

```python
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
```

```python
plt.imshow(closing, cmap='gray')
plt.axis('off')
```

```python
imgero = np.hstack((img, closing))
plt.imshow(imgero, cmap='gray')
plt.axis('off')
```

## Gradient

```python
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)
```

```python
plt.imshow(gradient, cmap='gray')
plt.axis('off')
```

```python
imgero = np.hstack((img, gradient))
plt.imshow(imgero, cmap='gray')
plt.axis('off')
```
