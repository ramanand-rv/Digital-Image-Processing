# Experiment 2

```python
#Importing required libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

## 1. Reading Image and Displaying

```python
# To read Image
img = cv2.imread("photo.jpg")
```

```python
# Since OpenCv reads image in BGR while camera captures in RGB format
# Hence we convert colors from BGR to RGB in order to dispay original colors 
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
# To Display Image
plt.imshow(img)
```

## 2. Display Informations
### a) Size in bits
### b) Row * Column

```python
# To find size of the image in Bytes
size = os.path.getsize("photo.jpg")

# Size of the image in bits
print("Image size in bits: ", size*8, "Bits")
```

```python
# To find Dimension (Row * Column) of the image
row = img.shape[0]
column = img.shape[1]
channels = img.shape[2]

# Displaying Row and Column value
print("Row * Column * channels :", row,"*", column, "*", channels)
```

## 3. Display RGB channels seperately

```python
## Displaying R channel image
R = img.copy()
R[:, :, 1] = 0
R[:, :, 2] = 0
# plt.imshow(R)
```

```python
# Displaying G channel image
G = img.copy()
G[:, :, 0] = 0
G[:, :, 2] = 0
# plt.imshow(G)
```

```python
# Displaying B channel image
B = img.copy()
B[:, :, 0] = 0
B[:, :, 1] = 0
# plt.imshow(B)
```

```python
# Displaying RGB channel separately
RGB = np.concatenate((R, G, B), axis= 1)
plt.imshow(RGB)
```

## 4. Displaying Images with 2 color channels combined

```python
# Displaying RG channel image
RG = img.copy()
RG[:, :, 2] = 0
# plt.imshow(RG)
```

```python
# Displaying GB channel image
GB = img.copy()
GB[:, :, 0] = 0
# plt.imshow(GB)
```

```python
# Displaying RB channel image
RB = img.copy()
RB[:, :, 1] = 0
# plt.imshow(RB)
```

```python
# Displaying RG, GB, RB channel images separately
RGB_2 = np.concatenate((RG, GB, RB), axis=1)
plt.imshow(RGB_2)
```

## 5. Color image to Gray image

```python
# Converting Color image to gray image
imgGray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

# Displaying Gray image
plt.imshow(imgGray, cmap='gray')
plt.imsave('imgGray.jpg',imgGray, format = 'jpg' )
```

## 6. Display Information

```python
# Finding dimensions of image
rowG, columnG = imgGray.shape[0], imgGray.shape[1]

# Displaying rows and columns
print("Rows * Columns :", rowG, "*", columnG)
```

```python
# Finding size of the image
sizeG = os.path.getsize("imgGray.jpg")

# Displaying the size of image
print("Size of the image :", sizeG*8, "bits")
```

## 7. Gray image to Black and White image

```python
# Gray image to black and white image
(thresh, imgBW) = cv2.threshold(imgGray, 150, 255, cv2.THRESH_BINARY)

# Displaying black & white image
plt.imshow(imgBW, cmap='gray')
plt.imsave('imgBW.jpg', imgBW, format = 'jpg')
```

## 8. Display information

```python
# Finding dimensions of the image
rowBW = imgBW.shape[0]
columnBW = imgBW.shape[1]

# Displaying rows and columns
print("Rows * Columns :", rowBW, "*", columnBW)
```

```python
# Finding the size of the image
sizeBW = os.path.getsize("imgBW.jpg")

# Displaying the size of the image
print("Size of the image :", sizeBW*8, "bits")
```

```python
# Displaying matrix of image
print(np.array(imgBW))
```
